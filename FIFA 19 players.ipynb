# Data Preprocessing

## Dropping all images

```python
data_no_img = data.drop(columns=["Photo", "Flag", "Club Logo", "Real Face"], axis=1)
```

```python
data_imp = data_no_img.drop(columns=["ID", "Nationality", "Club", "Preferred Foot",
                                     "Loaned From", "Joined", "Jersey Number",
                                     "Contract Valid Until", "Name", "Work Rate",
                                     "Body Type", "Position", "Height", "Weight", 
                                     "LS", "ST", "RS", "LW", "LF", "CF", "RF", "RW",
                                     "LAM", "CAM", "RAM", "LM", "LCM", "CM", "RCM", 
                                     "RM", "LWB", "LDM", "CDM", "RDM","RWB", "LB",
                                     "LCB", "RCB", "RB", "CB"], axis=1)
```

```python
data_imp = data_imp.fillna(0)
```

```python
data_imp.info()
```

## Convert money amount to float number
for Value, Wage, and Release Claues

```python
def convert_money(money):
    if money == 0:
        return float(0)
    else:
        money = money[1:] # remove "€" symbol
        unit = money[-1]
        if unit == 'K':
            value = float(money[0:-1])*1000
        elif unit == 'M':
            value = float(money[0:-1])*1000000
        else:
            value = float(money)
        return value
```

```python
convert_money("€565K")
```

```python
data_imp["Wage"] = data_imp["Wage"].apply(convert_money)
```

```python
data_imp["Value"] = data_imp["Value"].apply(convert_money)
```

```python
data_imp["Release Clause"] = data_imp["Release Clause"].apply(convert_money)
```

# Get Train and Test data

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(data_imp, test_size=0.2)
```

```python
train.info()
```

```python
y_train = train["Value"]
x_train = train.drop("Value", axis=1)
```

## Normalization

```python
x_train = (x_train-x_train.mean())/x_train.std()
```

```python
x_train.info()
```

```python
len(y_train)
```

```python
y_test = test["Value"]
x_test = test.drop("Value", axis=1)
```

## Normalization

```python
x_test = (x_test-x_test.mean())/x_test.std()
```

```python
x_test.info()
```

```python
len(y_test)
```

# Build NN model

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Flatten
```

```python
model = Sequential([
    Dense(43, input_shape=(43, )),
    Activation('relu'),
    Dense(33),
    Activation('relu'),
    Dense(23),
    Activation('relu'),
    Dense(13),
    Activation('relu'),
    Dense(11),
    Activation('relu'),
    Dense(1),
    Activation('relu')
])
```

```python
print(model.summary())
```

```python
from keras import optimizers
```

```python
# For a mean squared error regression problem
adam = optimizers.Adam(lr=0.003)
model.compile(optimizer=adam,
              loss='mse')
```

```python
# Train the model, iterating on the data in batches of 64 samples
model.fit(x_train, y_train, epochs=10, batch_size=64)
```

```python
model.evaluate(x_test, y_test, batch_size=64)
```
